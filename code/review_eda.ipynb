```python
df = pd.DataFrame(raw_json)
```

```python
df_chicken = df[df['text'].str.contains('chicken sandwich')]
```

```python
df_chicken.shape[0]
```

```python
df_chicken_stars = df_chicken.groupby('business_id').mean()
```

```python
df_chicken_count = df_chicken.value_counts('business_id').reset_index(name='count')
```

```python
df_chicken_final = pd.merge(df_chicken_stars, df_chicken_count, on='business_id', how='inner')
```

```python
df_chicken_final.loc[(df_chicken_final['count'] >= 10)].sort_values(by=(['stars','count']),ascending=False)
```
